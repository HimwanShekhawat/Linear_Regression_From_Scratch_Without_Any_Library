# Linear Regression from Scratch

This notebook demonstrates linear regression using gradient descent implemented from scratch in Python.

## 📌 Step 1: Import Libraries
We begin by importing necessary libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 📊 Step 2: Generate Synthetic Dataset
We generate a random linear dataset using the formula $y = 4 + 3x + \text{noise}$.

```python
# Set seed for reproducibility
np.random.seed(42)

# Generate synthetic data
num_samples = 100
x_train = 2 * np.random.rand(num_samples, 1)
true_slope = 3
true_intercept = 4
noise = np.random.randn(num_samples, 1)
y_train = true_intercept + true_slope * x_train + noise
```

```python
# Plot the data
plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color="blue", label="Training data", alpha=0.7)
plt.title("Dummy Linear Regression Data")
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()
```

## 🧮 Step 3: Define Cost Function
This function calculates the Mean Squared Error (MSE) between predictions and actual values.

```python
def compute_cost(x, y, w, b):
    m = len(x)
    predictions = w * x + b
    errors = predictions - y
    cost = (1 / (2 * m)) * np.sum(errors ** 2)
    return cost
```

## 🔁 Step 4: Gradient Descent Algorithm
We update weights `w` and `b` iteratively using the gradients to minimize the cost.

```python
def gradient_descent(x, y, w, b, alpha, iterations):
    m = len(x)
    cost_history = []

    for i in range(iterations):
        predictions = w * x + b
        errors = predictions - y

        dj_dw = (1 / m) * np.sum(errors * x)
        dj_db = (1 / m) * np.sum(errors)

        w -= alpha * dj_dw
        b -= alpha * dj_db

        cost = compute_cost(x, y, w, b)
        cost_history.append(cost)

        if i % 100 == 0:
            print(f"Iteration {i}: Cost {cost:.4f}, w = {w:.4f}, b = {b:.4f}")

    return w, b, cost_history
```

## 🏋️ Step 5: Train the Model
Let's initialize the parameters and run gradient descent.

```python
# Initialize parameters
initial_w = 0
initial_b = 0
learning_rate = 0.1
iterations = 1000

# Train the model
w_final, b_final, cost_history = gradient_descent(x_train, y_train, initial_w, initial_b, learning_rate, iterations)
```

## 📈 Step 6: Plot the Regression Line

```python
# Plot the regression line
plt.figure(figsize=(8, 5))
plt.scatter(x_train, y_train, color="blue", label="Training data", alpha=0.6)
plt.plot(x_train, w_final * x_train + b_final, color="red", label="Regression Line")
plt.title("Linear Regression Result")
plt.xlabel("x_train")
plt.ylabel("y_train")
plt.legend()
plt.grid(True)
plt.show()

print(f"Final Weights: w = {w_final:.4f}, b = {b_final:.4f}")
```
